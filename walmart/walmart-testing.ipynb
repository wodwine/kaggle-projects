```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import warnings

warnings.filterwarnings('ignore')
```

#### Import data

```python
data = pd.read_csv("data/walmart-preprocessing-test.csv")
data
```

#### Define X

```python
X = data.drop(["DateTime"],axis=1)
X
```

#### Load model

```python
import pickle

loaded_model = pickle.load(open("models/RandomForestModel.sav", 'rb'))
y = loaded_model.predict(X)
```

```python
predicted_list = []
for i,row in enumerate(data.values,0):
    Id = f"{str(row[0])}_{str(row[1])}_{str(row[3])}"
    predicted_list.append([Id,y[i]])
predicted_df = pd.DataFrame(data=predicted_list,columns=["Id","Weekly_Sales"])
predicted_df
```

```python
predicted_df.to_csv("outputs/predicted-test.csv")
```
