```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import warnings

warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv("data/walmart-data.csv")
data
```

```python
data["Type"] = data["Type"].map({"A":3,"B":2,"C":1})
```

```python
plt.scatter(data["Size"],data["Weekly_Sales"],c=data["Type"]) # Yellow: A, Green: B, Purple: C
plt.xlabel("Size")
plt.ylabel("Weekly Sales")
plt.show()
```

```python
stores_group = data.groupby(['Store']).mean()
stores_group.head()
```

```python
plt.scatter(stores_group["Size"],stores_group["Weekly_Sales"],c=stores_group["Type"]) # Yellow: A, Green: B, Purple: C
plt.xlabel("Size")
plt.ylabel("Weekly Sales")
plt.show()
```
